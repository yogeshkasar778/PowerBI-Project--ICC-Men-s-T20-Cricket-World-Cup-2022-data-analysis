### Data cleaning and Transformation in python pandas

```python
#import necessary library
import pandas as pd
import json
```

```python
with open('t20_wc_match_results.json') as f:
    data=json.load(f)
match_data=pd.DataFrame(data[0]['matchSummary'])
match_data.head()
```

```python
match_data.shape
```

```python
match_data.rename({'scorecard':'matchID'}, axis=1, inplace=True)
match_data.head()
```

```python
matchID_dict={}

for index, row in match_data.iterrows():
    key1=row['team1'] + ' Vs ' + row['team2']
    key2=row['team2'] + ' Vs ' + row['team1']
    
    matchID_dict[key1] = row['matchID']
    matchID_dict[key2] = row['matchID']
    
matchID_dict
```

```python
match_data.to_csv('E:\\Data Science\\Power BI\\match_data.csv',index=False)
```

**Batting summary**

```python
with open('t20_wc_batting_summary.json') as f:
    data=json.load(f)
    
    all_records=[]
    for rec in data:
        all_records.extend(rec['battingSummary'])
batting_data=pd.DataFrame(all_records)
batting_data.head()
```

```python
batting_data["Out/Not_Out"]=batting_data.dismissal.apply(lambda x: "Out" if len(x) > 0 else "Not_Out")
batting_data.head()
```

```python
batting_data.drop(columns=["dismissal"], inplace=True)
batting_data.head(10)
```

```python
batting_data['batsmanName']=batting_data['batsmanName'].apply(lambda x: x.replace('â€',''))
batting_data['batsmanName']=batting_data['batsmanName'].apply(lambda x: x.replace('\xa0',''))
batting_data.head(11)                                                               
```

```python
batting_data["matchID"]=batting_data["match"].map(matchID_dict)
batting_data.head()
```

```python
batting_data.shape
```

```python
batting_data.to_csv('E:\\Data Science\\Power BI\\batting_summary.csv',index=False)
```
